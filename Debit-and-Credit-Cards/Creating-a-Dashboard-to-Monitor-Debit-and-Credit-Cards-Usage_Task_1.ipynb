Once we have downloaded and set up Chrome Webdriver, we can proceed with automating the download of the Excel files from the Reserve Bank of India website.

```python
url = " https://www.rbi.org.in/Scripts/ATMView.aspx"
chrome_driver_path = "C:\\Users\\birar\\Downloads\\chromedriver_win32"
webdriver.chrome.service.Service.executable_path = chrome_driver_path
driver=  webdriver.Chrome( )
driver.get(url)
```

##### Here, click the 2022 link to get the access to the website which contains all the 2022 excel files

```python
html_file= driver.page_source
```

#### After using selenium, we can use beautifulsoup to extract the files from the website.

```python
data = BeautifulSoup(html_file)
```

```python
#Find all the elements which having "a" tag
data.find_all("a")
```

```python
# Find all <a> tags and print the href attribute
links = data.find_all("a")
for link in links:
    href = link.get("href")
    print(href)
```

### Finding all the links ending with .XLSX

```python
starting_URL="https://rbidocs.rbi.org.in/rdocs/ATM/DOCs/"
download_links = list(filter(lambda link: link.has_attr("href") 
                             and link["href"].startswith(starting_URL), 
                             data.find_all("a")))
```

```python
download_links = [link["href"] for link in download_links]
```

```python
download_links
```

<b>We want only the files till April 2022 </b>

```python
final_links = download_links[:9]
final_links
```

### Rename the file and download it

```python
name=["Dec_2022", "Nov_2022","Oct_2022", "Sept_2022", "Aug_2022", "Jul_2022", "Jun_2022", "May_2022", "Apr_2022"]
```

```python
[open(filename + ".xlsx", "wb").write(requests.get(final_links).content) for final_links, filename in zip(final_links, name)]
```

### Check the files have download or not

```python
import os

# Get specific files from the source directory
file_extensions = [".xlsx"]  # Specify the file extensions you want

file_list = [file for file in os.listdir() if os.path.isfile(os.path.join(file)) and any(file.endswith(ext) for ext in file_extensions)]

# Print the specific files
for file in file_list:
    print(file)
```

## Now, download file from January 2023 to March 2023

```python
from selenium import webdriver

options = webdriver.ChromeOptions()
options.add_argument("--headless")  # Optional: Run Chrome in headless mode

driver = webdriver.Chrome(options=options)
driver.get(url)
```

```python
html_file_1= driver.page_source
soup1=BeautifulSoup(html_file_1)
driver.close()
```

```python
start_of_link="https://rbidocs.rbi.org.in/rdocs/ATM/DOCs/"
start_of_link
```

```python
download_URL = [link1["href"] for link1 in soup1.find_all("a", href=lambda href: href and href.startswith(start_of_link) and href.endswith(".XLSX"))]
```

```python
download_URL
```

### We want file from Jan 2023 to mar 2023

```python
dl2=download_URL[2:5]
dl2
```

```python
name1=["Mar_2023","Feb_2023","Jan_2023"]
```
